```python
import json
```

```python
# Lets write a function to load the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

```python
# Write a function to display user and their connections
def display_users(data):
    print("User and their connection:")
    for user in data['users']:
        print(f"ID:{user['id']}- {user['name']} is friend with: {user['friends']} and like pages are {user['liked_pages']}")
    print("\npages Information:")
    for page in data['pages']:
        print(f"{page['id']}: {page['name']}")

display_users(data)
```
